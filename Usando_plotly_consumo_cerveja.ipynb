```python
import plotly.plotly as py
import pandas as pd
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot,iplot
import plotly.graph_objs as go
#import all the necessary libraries above
#Acima import todas as bibliotecas necessárias
```

```python
init_notebook_mode(connected = True)
#initiate the ini-notebook
#aqui inicio o initi_notebook
```

```python
#
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')
```

```python
df.head()
```

```python
df2 = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_beer_consumption_per_capita')
```

```python
countries_beer = df2[0]
countries_beer.head()
```

```python
countries_beer[1] = countries_beer[1].str.replace('[', '')
countries_beer[1] = countries_beer[1].str.replace(']', '')
countries_beer[1] = countries_beer[1].str.replace('2', '')
countries_beer[1] = countries_beer[1].str.replace('3', '')
countries_beer[1] = countries_beer[1].str.replace('4', '')
countries_beer[1] = countries_beer[1].str.replace('5', '')
countries_beer[1] = countries_beer[1].str.replace('6', '')
countries_beer[1] = countries_beer[1].str.replace('7', '')
countries_beer[1] = countries_beer[1].str.replace('8', '')
countries_beer[1] = countries_beer[1].str.replace('9', '')
countries_beer[1] = countries_beer[1].str.replace('0', '')
countries_beer[1] = countries_beer[1].str.replace('10', '')
countries_beer[1] = countries_beer[1].str.replace('11', '')  
countries_beer[1] = countries_beer[1].str.replace('1', '')
```

```python
countries = df
countries.head()
```

```python
countries_beer.columns = countries_beer.iloc[0] 
countries_beer.reindex(countries_beer.index.drop(0))
```

```python
countries_concatenated = pd.merge(left=countries, right=countries_beer, left_on='COUNTRY', right_on='Country')

```

```python
countries_concatenated = countries_concatenated[['COUNTRY','CODE','Rank','Consumption per capita [1] (litres)']]
countries_concatenated.head()
```

```python
countries_concatenated.info()
```

```python
countries_concatenated['Rank'] = countries_concatenated['Rank'].astype(int)
```

```python
countries_concatenated.sort(columns='Rank', axis=0, ascending=True)
countries_concatenated.set_index('Rank', drop=True, append=False, inplace=True, verify_integrity=False)

```

```python
countries_concatenated.sort_index(axis=0, level=None, ascending=True, inplace=True)
```

```python
countries_concatenated
```

```python
data = dict(type = 'choropleth',
           locations = countries_concatenated['CODE'],
           z = countries_concatenated['Consumption per capita [1] (litres)'],
            text = countries_concatenated['COUNTRY'],
            colorbar = {'title': 'Consumption per capita [1] (litres)'})
```

```python
layout = dict(title = 'Consumption of beer - Per Capita',
             geo = dict(showframe = False,
                       projection = {'type':'natural earth'}))
```

```python
choromap3 = go.Figure(data=[data],layout=layout)
```

```python
iplot(choromap3)
```
